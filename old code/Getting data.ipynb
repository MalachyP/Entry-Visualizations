```python
import time

COLUMN_LOAD_TIME = 0.4

def get_column_names(driver):
    # assume the first column exists (natural)
    column_names = []
    prev_column = None

    # keep checking if change to column
    next_column = get_text(driver)
    while(prev_column != next_column):
        column_names.append(next_column)
        prev_column = next_column

        # move to the next column cell and wait for the page to load
        move_direction(driver, right=1)
        time.sleep(COLUMN_LOAD_TIME)

        # get the next column
        next_column = get_text(driver)
    
    # move back to the start
    move_direction(driver, left=len(column_names))

    return column_names
```

```python
import numpy as np

def get_frame(rows, driver):
    column_names = get_column_names(driver)
    move_direction(driver, down=1)

    # initialize the data frame
    df_out = pd.DataFrame(data = np.zeros((rows, len(column_names))), columns = column_names)

    for column_name in column_names:
        # set the new values
        column_values = get_column(rows, driver)
        df_out[column_name] = pd.Series(data = column_values, name = column_name)

        # move all the way back up and to the right
        move_direction(driver, right=1, up=rows)

        print(f"{time.strftime('%H:%M:%S', time.localtime())} {column_name}")
    
    return df_out   
```

```python
TOP_TITLE = "border-top-width"
BOTTOM_TITLE = "border-bottom-width"

PADDING = 0

def get_cell_coordinates(driver):
    coordinates_dict = {}
    
    # get the pixel ratio (used for the image)
    pixel_ratio = driver.execute_script("return window.devicePixelRatio")
    
    # get the coordinates of the current cell
    active_cell_width_elements = driver.find_elements(By.CSS_SELECTOR, 'div.range-border.active-cell-border')

    # ensure there is a top, bottom, left and right
    if (len(active_cell_width_elements) != 4):
        print("didn't get four elements")
        exit()

    # find the bottom and top elements
    for i in range(4):
        # get the current style
        curr_style = active_cell_width_elements[i].get_attribute('style')

        # find the top and bottom elements
        if (TOP_TITLE in curr_style):
            active_cell_width_top_element = active_cell_width_elements[i]
        elif (BOTTOM_TITLE in curr_style):
            active_cell_width_bottom_element = active_cell_width_elements[i]

    # assign values to the dictionary
    coordinates_dict["left"] = active_cell_width_top_element.location["x"] * pixel_ratio + PADDING
    coordinates_dict["right"] = (active_cell_width_top_element.location["x"] + active_cell_width_top_element.size["width"]) * pixel_ratio - PADDING
    coordinates_dict["top"] = active_cell_width_top_element.location["y"] * pixel_ratio + PADDING
    coordinates_dict["bottom"] = (active_cell_width_bottom_element.location["y"] + active_cell_width_bottom_element.size["height"]) * pixel_ratio - PADDING

    return coordinates_dict
```

```python
from selenium.webdriver.common.action_chains import ActionChains
import time

RESET_WAIT = 0.1

def move_direction(driver, right=0, left=0, up=0, down=0, fast=False):
    action = ActionChains(driver)

    if (not fast):
        wait = RESET_WAIT
    else:
        wait = 0
        
    for _ in range(right):
        action.send_keys(Keys.ARROW_RIGHT)
        action.perform()
        time.sleep(wait)

    for _ in range(left):
        action.send_keys(Keys.ARROW_LEFT)
        action.perform()
        time.sleep(wait)

    for _ in range(up):
        action.send_keys(Keys.ARROW_UP)
        action.perform()
        time.sleep(wait)

    for _ in range(down):
        action.send_keys(Keys.ARROW_DOWN)
        action.perform()
        time.sleep(wait)
```

#### Coordinates functions

```python
def get_integer_batch_coordinates(column_coordinates, batch_rows):
    cell_coordinates = []

    # get the ranges of the column
    top = column_coordinates["top"]
    y_range = column_coordinates["bottom"] - top
    
    for i in range(batch_rows):
        # get the new cell
        curr_cell_coordinates = column_coordinates.copy()
        curr_cell_coordinates["top"] = top + round(i * (y_range / batch_rows))
        curr_cell_coordinates["bottom"] = top + round((i + 1) * (y_range / batch_rows))
        
        cell_coordinates.append(curr_cell_coordinates)

    return cell_coordinates
    
```

```python
"""will make it relative to the cropped image """


""" will output a dictionary of coordinates for each column
if index is not an integer, will store a column (coordinate unit)
else it will store a list of coordinate units
Note: will finish in the bottom left"""
def get_column_batch_coordinates(column_indices, column_integer_indices, batch_rows, driver):   
    # create the new coordinates    
    column_coordinates_dict = {}
    
    # get the top and bottom coordinates
    top_coordinates = get_cell_coordinates(driver)
    move_direction(driver, down=batch_rows-1)
    bottom_coordinates = get_cell_coordinates(driver)

    # get the template coordinate and left_marker
    left_marker = top_coordinates["left"]

    # get each column coordinates
    prev_index = 0    
    for column_index in column_indices:
        # move to the next column
        move_direction(driver, right=column_index - prev_index)
        prev_index = column_index
        
        # get the entire column coordinates and standardize
        curr_column_coordinates = get_cell_coordinates(driver)
        curr_column_coordinates["left"] = curr_column_coordinates["left"] - left_marker
        curr_column_coordinates["right"] = curr_column_coordinates["right"] - left_marker
        curr_column_coordinates["top"] = 0
        curr_column_coordinates["bottom"] = bottom_coordinates["bottom"] - top_coordinates["top"]

        # deal with integers if necessary
        if (not column_index in column_integer_indices):
            column_coordinates_dict[column_index] = curr_column_coordinates

        else:
            column_coordinates_dict[column_index] = get_integer_batch_coordinates(curr_column_coordinates, batch_rows)

    # get the ranges
    column_coordinates_dict["y range"] = bottom_coordinates["bottom"] - top_coordinates["top"]
    column_coordinates_dict["x range"] =  curr_column_coordinates["right"]
    
    # reset to the former position
    move_direction(driver, left=column_indices[-1], up=batch_rows-1)

    return column_coordinates_dict
```

```python
from io import BytesIO, StringIO
import numpy as np

import pytesseract

pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
INTEGER_CONFIG = f'--psm 6 -c tessedit_char_whitelist=0123456789.'

PADDING = 2

def get_text_from_image(coordinates, img, batch_rows, integer=False):       
    # crop the column
    img = img.crop((coordinates["left"] + PADDING, coordinates["top"] + PADDING, 
                    coordinates["right"] - PADDING, coordinates["bottom"] - PADDING))
    
    # handling column data
    if (not integer):
        
        # get the data frame results
        result_df = pytesseract.image_to_data(img, output_type=pytesseract.Output.DATAFRAME)
        
        # find the nearest line
        result_df.loc[:, "index"] = result_df["top"].apply(lambda x: round(batch_rows * (x / img.size[1])))
        result_df = result_df.dropna(subset="text")
        result_df = result_df.groupby("index")["text"] \
                             .apply(lambda texts: " ".join([str(x) for x in texts]).strip())

        # will in the missing values
        result = list(np.zeros(batch_rows))
        for index, text in result_df.to_dict().items():
            # ensure not picking up random text
            if (0 <= index and index < batch_rows):
                result[index] = text

    # getting a single float value
    else:
        result = pytesseract.image_to_string(img, config=INTEGER_CONFIG)
        result = result.strip()
    
    return result
```

```python
from PIL import Image

def get_relative_image(column_coordinates_dict, driver):
    # get the screen shot
    png = driver.get_screenshot_as_png()

    # get the coordinates of the relative image
    curr_cell_coordinates = get_cell_coordinates(driver)
    relative_image_coordinates = (curr_cell_coordinates["left"], curr_cell_coordinates["bottom"] - column_coordinates_dict["y range"],
                                  curr_cell_coordinates["left"] + column_coordinates_dict["x range"], curr_cell_coordinates["bottom"])
    
    # get the relative image
    img = Image.open(BytesIO(png))
    img = img.crop(relative_image_coordinates)

    return img
```

```python
def get_column_batch_text(column_coordinates_dict, column_integer_indices, batch_rows, driver, indices_valid=False):
    # result storage
    batch_results = []

    # getting the relative image
    img = get_relative_image(column_coordinates_dict, driver)

    # get the result from each (make sure to avoid x range and stuff)
    for column_index, column_coordinates in list(column_coordinates_dict.items())[:-2]:
        # deal with column data
        if (not column_index in column_integer_indices):
            batch_results.append(get_text_from_image(column_coordinates, img, batch_rows))

        # deal with integers, but no ongoing rows
        elif (not indices_valid):
            curr_column_results = []
            for cell_coordinates in column_coordinates:
                curr_column_results.append(get_text_from_image(cell_coordinates, img, batch_rows, integer=True))

            batch_results.append(curr_column_results)

        # deal with results
        else:
            # get the new results, making sure to skip invalid rows
            curr_column_results = []
            for i in range(batch_rows):
                if (not i in indices_valid):
                    # if not valid don't bother
                    curr_column_results.append("")
                    
                else:
                    # if valid, get the result
                    curr_column_results.append(get_text_from_image(column_coordinates[i], img, batch_rows, integer=True))

            # remove all the new indices that are blank
            for i, value in enumerate(curr_column_results):
                if ((i in indices_valid) and (value == "")):
                    indices_valid.remove(i)

            # save the results
            batch_results.append(curr_column_results)

    return batch_results
```

#### Resetting Functions

```python
RESET = 10
ERROR_MARGIN = 5

""" will makes sure everything is in present view. Tries to prevent hitting the end of the list
Note: assumes no repeats in column names"""
def reset_window(driver):
    # count the number of times ACTUALLY moved
    count = 0

    # get the last cell value
    last_cell_value = get_displayed_text(driver)
    
    while (count < RESET):
        # move to the next cell
        move_direction(driver, right=1)

        # get the new cell value
        new_cell_value = get_displayed_text(driver)
    
        # check that we did actually move
        if (new_cell_value == last_cell_value):
            break
        
        # update the count
        count += 1

        # if an empty column, need to move back (otherwise will keep getting duplicates)
        if (not new_cell_value):
            break
        
        # update the last right cell
        last_cell_value = new_cell_value

    # move back count times
    move_direction(driver, left=count)
```

```python
""" inputs:
- num_cols: is the number of columns in this column batch
    outputs:
- a 3-D list. The first dimension will be all the batches. The second dimension will be each column. The third will be the 
  actual values within that column batch, will be organised later
Note:
- assumes starting on the correct cell (in the first value)
- assumes that the number of rows is known"""
```

```python
import time
import numpy as np

BATCH_ROWS = 14

def get_columns(curr_col_batch, database_dict, driver):
    # get the current column indices and integer indices
    column_indices = database_dict["divisions"][curr_col_batch]    
    column_integer_indices = list(set(column_indices) & set(database_dict["integers"]))

    # make the indices relative
    column_integer_indices = [x - column_indices[0] for x in column_integer_indices]
    column_indices = [x  - column_indices[0] for x in column_indices]

    # include any emergency integers
    if (database_dict.get("integer flag") and curr_col_batch in database_dict["integer flag"].keys()):
        column_integer_indices.extend(database_dict["integer flag"][curr_col_batch])

    # the output as specified in the doc string
    columns_results = []

    # get the initial relative coordinates
    column_coordinates_dict = get_column_batch_coordinates(column_indices, column_integer_indices, BATCH_ROWS, driver)
    move_direction(driver, up=1)
    
    # loop for each iteration
    iterations = int(np.ceil(database_dict["rows"] / BATCH_ROWS))
    for i in range(iterations):
        print(f"{get_time()} database: {database_dict['name']} column_batch: {curr_col_batch} batch: {i+1}")

        # move to get the next batch
        move_direction(driver, down=BATCH_ROWS)
        
        # get the text
        batch_results = get_column_batch_text(column_coordinates_dict, column_integer_indices, BATCH_ROWS, driver)
        
        # add to the values
        columns_results.append(batch_results)
        
    # get ready for the next columns
    move_direction(driver, up=iterations * BATCH_ROWS + 1, fast=True)

    return columns_results
```

```python
def unravel_columns(batch_results):
    # initialize with the first batch
    columns_results = batch_results[0]

    # get each batch
    for batch in batch_results[1:]:
        for i in range(len(columns_results)):
            columns_results[i].extend(batch[i])

    return columns_results        
```

```python
import numpy as np

RESULT_BATCHES = 7   # how many (uni, gpa, gamsat) sections there are
BATCH_ROWS = 14   # how many rows there are per batch
BATCH_COLS = 3    # supposed to be three because three of (uni, gpa, gamsat)
COLUMN_LOAD = 0.2    # the wait time before going down a column

def get_results(database_dict, driver):    
    # number of iterations
    iterations = int(np.ceil(database_dict["rows"] / BATCH_ROWS))

    # the final results
    results = []
    column_indices = [0, 1]
    column_integer_indices = [1]

    # initialize rows valid
    rows_valid = []
    for i in range(iterations):
        rows_valid.append(list(range(BATCH_ROWS)))

    interview_column = True

    # repeat for each set of columns
    for _ in range(RESULT_BATCHES):
        columns_results = []

        # get the initial relative coordinates
        column_coordinates_dict = get_column_batch_coordinates(column_indices, column_integer_indices, BATCH_ROWS, driver)
        move_direction(driver, up=1)
        
        for i in range(iterations):
            print(f"{get_time()} database: {database_dict['name']} results batch: {_*iterations + i + 1}")
            
            # move to get the next batch
            move_direction(driver, down=BATCH_ROWS)
            
            # get the text
            if (interview_column):
                batch_results = get_column_batch_text(column_coordinates_dict, column_integer_indices, BATCH_ROWS, driver)
            else:
                batch_results = get_column_batch_text(column_coordinates_dict, column_integer_indices, BATCH_ROWS, driver, indices_valid=rows_valid[i])
            
            # add to the values
            columns_results.append(batch_results)
        
        # get ready for the next batch
        move_direction(driver, right=BATCH_COLS, up=iterations*BATCH_ROWS + 1, fast=True) # + 1 because of get_read_next_batch
        reset_window(driver)
        move_direction(driver, down=1)

        # add the batches
        results.append(columns_results)

        # make sure not in interview
        interview_column = False

    # move back to the original position
    move_direction(driver, left=BATCH_COLS * RESULT_BATCHES, fast=True)

    return results
```

```python
def unravel_results(results):
    column_values = []
    
    # get each result batch index
    for results_batch_index in range(RESULT_BATCHES):
        curr_unis = []
        curr_gamsat = []

        # iterate through each batch
        for batch in results[results_batch_index]:
            curr_unis.extend(batch[0])
            curr_gamsat.extend(batch[1])
        column_values.append(curr_unis)
        column_values.append(curr_gamsat)

    return column_values
```

```python
COLUMN_LOAD = 0.2

def get_dataframe_advanced(database_dict, driver):
    # convient names
    divisions = database_dict["divisions"]
    result_cols_start = database_dict["result cols"][0]
    
    # initializing for loop
    visited_results = False
    curr_col_batch = 0
    curr_col_index = 0    
    all_values = []
    
    # basically keep iterating until no more columns batches to do and visited results
    while ( (not visited_results) or (curr_col_batch < len(divisions)) ):
        # check if need to visit results and if either have no more batches, or the next batch is after results
        if ( (not visited_results) and (curr_col_batch >= len(divisions) or divisions[curr_col_batch][0] > result_cols_start) ):           
            # move into the correct position
            move_direction(driver, right=result_cols_start - curr_col_index, fast=True)
            reset_window(driver)
            move_direction(driver, down=1)
            
            # update the current index
            curr_col_index = result_cols_start
            
            # get the results
            results = get_results(database_dict, driver)
            all_values.extend(unravel_results(results))

            # move back to the columns
            move_direction(driver, up=1)
            
            # update
            visited_results = True

        # check if the next col_batch is before the results section
        else:           
            # move into the correct position
            move_direction(driver, right=divisions[curr_col_batch][0] - curr_col_index, fast=True)
            reset_window(driver)
            move_direction(driver, down=1)

            # get the results
            results = get_columns(curr_col_batch, database_dict, driver)
            all_values.extend(unravel_columns(results))
            
            # update the current index variables
            curr_col_index = divisions[curr_col_batch][0]
            curr_col_batch += 1

            # move up
            move_direction(driver, up=1)

    # create the data frame (in doc string)
    df_out = pd.DataFrame(data=np.array(all_values).T, columns=[str(x) for x in range(len(all_values))])

    # keep only the amount of rows needed
    df_out = df_out.loc[np.array(range(database_dict["rows"]))]
    
    return df_out
```

```python
import numpy as np

ERROR_MARGIN = 5

""" BEHAVIOUR: will keep taking columns for the batches. Will stop if:
    - get to the results section
    - makes a column out of view
    Will automatically skip the skipped columns. Will still mean they may be screen shotted if they are in the centre however
    - Will also not begin on a skipped column
    For results section WILL ONLY INCLUDE the university and GPA (as can't recover this one)"""
def get_column_divisions(database_dict):
    hit_end_flag = False
    divisions = []
    strides = []
    visited_column_names = []
    curr_index = 0

    # start in the position of a new index while the end hasn't been hit, or haven't gone back to a visted column
    while ((not hit_end_flag) and (not get_displayed_text(driver) in visited_column_names)):       
        # initialize
        last_right = get_cell_coordinates(driver)["left"]
        start_index = curr_index
        curr_indices = []
        
        # while the left of the new cell has the same coordinates of the right of the last cell
        # NOTE: assumes starting on a column not meant to be skipped
        while (np.abs(get_cell_coordinates(driver)["left"] - last_right) < ERROR_MARGIN):
            print(f"curr index is {curr_index}")
            
            # make sure that it isn't an empty column as well
            if (not get_displayed_text(driver)):
                hit_end_flag = True
                print(f"broke as no displayed text for curr index {curr_index}")
                break

            # make sure that the cell hasn't hit a results column (will only hit the start)
            if (curr_index in database_dict["result col"][0]):
                break

            # now it is safe to include this column in the batch (will include rows meant to skip
            visisted_column_names.append(get_displayed_text(driver))
            curr_indices.append(curr_index)

            # update everything
            curr_index += 1
            last_right = get_cell_coordinates(driver)["right"]

            # move to the next cell
            move_direction(driver, right=1)

        # find the number of valid values in this index
        stride = curr_index - start_index

        # ensure actual indicies have been included
        if (stride != 0):
            # append the information
            divisions.append(curr_indices)
            strides.append(stride)
            
            # get it so the next column is on the far left now
            get_ready_columns_batch(driver)

    # reset to the original position
    move_direction(driver, left=curr_index)

    return divisions        
```
